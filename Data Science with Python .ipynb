```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('Churn_Modelling.csv', parse_dates= True)
```

```python
df1.head()
```

```python
df1['H-L']=df1.CreditScore-df1.Age
df1['ESl-100000']=df1['CreditScore'].rolling(100).mean()
```

```python

ax=plt.axes(projection='3d')
```

```python
ax=plt.axes(projection='3d')
ax.scatter(df1.index, df1['H-L'], df1['ESl-100000'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_ylabel('ESl-100000')
ax.set_ylabel('10000PA')
plt.show()
```
